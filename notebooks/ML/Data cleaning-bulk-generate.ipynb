```python
def generate_picture(id_crypto, crypto_name): 
    print('id_crypto : ' + id_crypto)
    print('crypto_name : ' + crypto_name)
    # ------------------ LOAD DATA FROM DATABASE ------------------ #

    # TODO : Replace with info from config file
    connection = create_engine('postgresql://dbuser:algocryptos@localhost:9091/algocryptos')

    # get data with query
    squery = "select oh.open_price, oh.high_price, oh.low_price, oh.close_price, oh.volume_aggregated, re.reddit_subscribers, oh.timestamp \n"
    squery += 'from histo_ohlcv oh\n'
    squery += "left outer join social_stats_reddit_histo re on (oh.timestamp = date_trunc('day', re.timestamp) + '00:00:00' and re.id_cryptocompare = oh.id_cryptocompare)\n"
    squery += 'where oh.id_cryptocompare = ' + id_crypto + '\n' #7605
    squery += 'and oh.timestamp >= (select min(timestamp) from social_stats_reddit_histo where id_cryptocompare = ' + id_crypto + ')\n'
    #squery += "and oh.timestamp between '2018-04-08 02:00:00+00' and '2018-04-08 12:00:00+00'\n"
    squery += 'order by oh.timestamp asc\n'
    df = psql.read_sql_query(squery, connection)
    
    # ------------------ PRE-PROCESSING : Data quality ------------------ #
    df2 = df

    # --------------------------------
    # TIMESTAMP
    # --------------------------------

    # mandatory when different timezones in database (column not recognized as datetime). To be careful : format = UTC, no timezone
    df2['timestamp'] = pd.to_datetime(df2.timestamp, utc=True)

    # --------------------------------
    # REDDIT SUBSCRIBERS
    # --------------------------------

    # [DQ] interpolate for missing values in reddit subscribers
    df2.reddit_subscribers = df2.reddit_subscribers.interpolate()

    # [DQ] remove rows with missing values after interpolate 
    df2 = df2[df2.reddit_subscribers.notnull()]

    # [DQ] remove duplicate datetime
    df2 = df2[~df2.timestamp.duplicated(keep='first')]

    # --------------------------------
    # OHLCV
    # --------------------------------

    # [DQ] drop rows with missing values (OHLCV)
    df2 = df2.loc[(df2.open_price != 0.0) & (df2.high_price != 0.0) & (df2.low_price != 0.0) & (df2.close_price != 0.0) & (df2.volume_aggregated != 0.0)]

    # [DQ] set index
    df2.set_index('timestamp', inplace = True)

    # [DQ] remove outliers in all data
    df2 = remove_outliers(df2)

    # [DQ] resample with period 1H + interpolation for missing values
    df2 = df2.resample('1H').interpolate()
    print('shape after interpolate : ' + str(df2.shape))
    
    # figure
    fig1 = plt.figure(figsize=(15,15))
    df2.close_price.plot()
    plt.savefig(id_crypto + ".png", dpi=30, transparent=True)
    #df2.volume_aggregated.plot(secondary_y=True)
```
