```python
import pandas as pd
import numpy as np
df = pd.read_json(r"bdd.json")

# Size
df.shape
```

```python
df['time'] = pd.to_datetime(df['time'],unit='s')
df = df.set_index('time')
```

```python
%matplotlib inline
df.plot(y="close", color="Green")
```

```python
#Show columns types
df.dtypes
```

```python
# Delete rows with close=0
df = df[df.close != 0]
```

```python
# Describe datas
#df.describe
```

```python
# Change period
df.resample('3D').agg({'close': np.mean, 'high': np.max, 'low': np.min, 'open': np.mean, 'volumefrom': np.sum, 'volumeto': np.sum})
```

```python
# Close : %change mean(3j) / mean(30j)
df2 = df.truncate(before='2018-01-23', after='2018-02-23')
print('%change : ' + str(round((df2.close.tail(3).mean() - df2.close.tail(30).mean()) / df2.close.tail(30).mean() * 100, 2)))
```

```python
# Volumes : %change mean(3j) / mean(30j)
df2 = df.truncate(before='2018-01-23', after='2018-02-23')
print('%change : ' + str(round((df2.volumefrom.tail(3).mean() - df2.volumefrom.tail(30).mean()) / df2.volumefrom.tail(30).mean() * 100, 2)))
```
